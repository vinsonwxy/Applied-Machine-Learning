```python
# Question 2
# (a)
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.spatial import distance

documents = np.load("/Desktop/cs5785/hw3/science2k-doc-word.npy")
print(documents.shape)
print(documents[:10, :10])
```

```python
pca = PCA(n_components = 2)
documents_reduced = pca.fit_transform(documents)
print documents_reduced.shape
```

```python
plt.scatter(documents_reduced[:, 0], documents_reduced[:, 1])
```

```python
for k in range(2, 26):
    kmeans = KMeans(n_clusters=k).fit(documents)
    centers = kmeans.cluster_centers_
    
    pca = PCA(n_components = 2)
    centers_reduced = pca.fit_transform(centers)
    
    print centers.shape
    print centers_reduced.shape
    plt.figure()
    plt.scatter(documents_reduced[:, 0], documents_reduced[:, 1])
    plt.scatter(centers_reduced[:, 0], centers_reduced[:, 1], color = 'r')
    plt.show()
```

```python
titles = pd.read_table("/Desktop/cs5785/hw3/science2k-titles.txt", header = None).values
titles.shape
```

```python
k = 3
kmeans = KMeans(n_clusters=k).fit(documents)

for label in range(k):
    center = kmeans.cluster_centers_[label].reshape(1, -1)
    data_labeled = documents[np.where(kmeans.labels_ == label)]
    dist = distance.cdist(data_labeled, center)
    print titles[np.argsort(dist[:,0])][:10, 0]
```
