```python
import numpy as np
import matplotlib.pyplot as plt
import re

# preprocess the loss and iteration
with open('iter.txt','r') as f:
    content = f.readlines()
    a = content[2].split('\xe2\x80\xa8')
    iterations = a[0:len(a):3]
    iterations = map(float,iterations)
    
with open('loss.txt','r') as f:
    content = f.readlines()
    b = content[3].split('\xe2\x80\xa8') + content[4].split('\xe2\x80\xa8')
    losses =  ['0.02731582408409078'] + b[2:len(b):3]
    for i in range(len(losses)):
        losses[i] = re.sub('\t.\t', '', losses[i])
    losses = map(float, losses)
```

```python
plt.plot(iterations,losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('question (c) loss over 5000 iterations')
plt.show()
```

```python
# halve the learning rate every 1000 iterations

with open('iter_varyLoss.txt','r') as f:
    content = f.readlines()
    tmp = content[1].split('\xe2\x80\xa8')
    iter2 =  tmp[0:len(tmp):3]
    iter2 = map(float,iter2)
    
with open('loss_varyLoss.txt','r') as f:
    content = f.readlines()
    tmp = content[3].split('\xe2\x80\xa8') + content[4].split('\xe2\x80\xa8')
    loss2 = ['0.03394177533534439'] + tmp[2:len(tmp):3]
    for i in range(len(loss2)):
        loss2[i] = re.sub('\n', '', loss2[i])
    loss2 = map(float, loss2)
```

```python
plt.plot(iter2,loss2)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Question(d) When the learning rate is halved every 1000 iterations')
plt.show()
print 'The network converges to a lower loss in this case.'
```

```python
# prepocess data for 2 hidden layers
with open('iter_2layer.txt','r') as f:
    content = f.readlines()
    tmp = content[0].split('\xe2\x80\xa8')
    iter_2hl =  tmp[0:len(tmp):3]
    iter_2hl = map(int,iter_2hl)
#     print iter_2hl
    
with open('loss_2layer.txt','r') as f:
    content = f.readlines()
    tmp = content[0].split('\xe2\x80\xa8')
    loss_2hl = tmp[0:len(tmp):3]
    for i in range(len(loss_2hl)):
        loss_2hl[i] = re.sub('\t.\t', '', loss_2hl[i])
    loss_2hl = map(float, loss_2hl)
```

```python
# prepocess data for 3 hidden layers
with open('iter_3layer.txt','r') as f:
    content = f.readlines()
    tmp = "".join(content[3:]).splitlines()
    tmp = tmp[1:len(tmp):3]
    tmp2 = content[2].split('\xe2\x80\xa8')
    iter_3hl =  tmp2[0:len(tmp2):3] + tmp
    iter_3hl = map(int,iter_3hl)
    
with open('loss_3layer.txt','r') as f:
    content = f.readlines()
    tmp = content[2].split('\xe2\x80\xa8')
    loss_3hl = tmp[0:len(tmp):3]
    for i in range(len(loss_3hl)):
        loss_3hl[i] = re.sub('\n', '', loss_3hl[i])
    loss_3hl = map(float, loss_3hl)
```

```python
# prepocess data for 5 hidden layers
with open('iter_5layer.txt','r') as f:
    content = f.readlines()
    tmp = content[1].split('\xe2\x80\xa8')
    iter_5hl =  tmp[1:len(tmp):3]
    iter_5hl = map(int,iter_5hl)
    
with open('loss_5layer.txt','r') as f:
    content = f.readlines()
    tmp = content[2].split('\xe2\x80\xa8')
    loss_5hl = tmp[0:len(tmp):3]
    for i in range(len(loss_5hl)):
        loss_5hl[i] = re.sub('\n', '', loss_5hl[i])
    loss_5hl = map(float, loss_5hl)
```

```python
# plot the losses vs iterations
plt.plot(iter_5hl,loss_5hl)
plt.plot(iter_3hl,loss_3hl)
plt.plot(iter_2hl,loss_2hl)
plt.plot(iter_1hl,loss_1hl)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend(['5 hidden layer','3 hidden layers','2 hidden layers','1 hidden layer'],loc='upper right')
plt.show()

print 'From the plot, we can see that the accuracy drops significantly when the number of hidden layers decreases to 2.'
print 'The number of layers specified here only entails the hidden fully-connected layers, which exclude the input and output layer.'
```
